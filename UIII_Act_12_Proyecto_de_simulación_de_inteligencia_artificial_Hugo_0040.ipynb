```python
# -------- CLASE --------

class Analista:

    lista_analistas = []

    def __init__(self, nombre, software, base_datos, reportes, area, excel, certificaciones):
        self.nombre = nombre
        self.software = software
        self.base_datos = base_datos
        self.reportes = reportes
        self.area = area
        self.excel = excel
        self.certificaciones = certificaciones

    # -------- MÉTODOS --------

    @classmethod
    def agregar_analista(cls):
        print("\n--- Agregar Analista ---")
        nombre = input("Nombre: ")

        for a in cls.lista_analistas:
            if a.nombre.lower() == nombre.lower():
                print("❌ Ya existe ese analista\n")
                return

        software = input("Software de visualización: ")
        base_datos = input("Base de datos manejada: ")
        reportes = input("Reportes generados: ")
        area = input("Área de negocio: ")
        excel = input("Nivel de Excel: ")
        certificaciones = input("Certificaciones: ")

        nuevo = Analista(nombre, software, base_datos, reportes, area, excel, certificaciones)
        cls.lista_analistas.append(nuevo)

        print("✔ Analista agregado\n")


    @classmethod
    def listar_analistas(cls):
        print("\n--- Lista de Analistas ---")

        if len(cls.lista_analistas) == 0:
            print("No hay analistas registrados\n")
        else:
            print("{:<15} {:<15} {:<15} {:<15} {:<15} {:<10} {:<15}".format(
                "Nombre", "Software", "BD", "Reportes", "Área", "Excel", "Certificaciones"
            ))
            print("-" * 100)

            for a in cls.lista_analistas:
                print("{:<15} {:<15} {:<15} {:<15} {:<15} {:<10} {:<15}".format(
                    a.nombre, a.software, a.base_datos,
                    a.reportes, a.area, a.excel, a.certificaciones
                ))
            print()


    @classmethod
    def actualizar_analista(cls):
        print("\n--- Actualizar Analista ---")
        nombre = input("Nombre a actualizar: ")

        for a in cls.lista_analistas:
            if a.nombre.lower() == nombre.lower():
                print("Nuevo software:", end=" ")
                a.software = input()
                print("Nueva base de datos:", end=" ")
                a.base_datos = input()
                print("Nuevos reportes:", end=" ")
                a.reportes = input()
                print("Nueva área:", end=" ")
                a.area = input()
                print("Nuevo nivel de Excel:", end=" ")
                a.excel = input()
                print("Nuevas certificaciones:", end=" ")
                a.certificaciones = input()

                print("✔ Analista actualizado\n")
                return

        print("❌ No encontrado\n")


    @classmethod
    def borrar_analista(cls):
        print("\n--- Borrar Analista ---")
        nombre = input("Nombre a eliminar: ")

        for a in cls.lista_analistas:
            if a.nombre.lower() == nombre.lower():
                cls.lista_analistas.remove(a)
                print("✔ Analista eliminado\n")
                return

        print("❌ No encontrado\n")


# -------- MENÚ --------

while True:
    print("===== SISTEMA DE ANALISTA (POO) =====")
    print("1. Agregar analista")
    print("2. Listar analistas")
    print("3. Actualizar analista")
    print("4. Borrar analista")
    print("5. Salir")

    opcion = input("Selecciona una opción: ")

    if opcion == "1":
        Analista.agregar_analista()
    elif opcion == "2":
        Analista.listar_analistas()
    elif opcion == "3":
        Analista.actualizar_analista()
    elif opcion == "4":
        Analista.borrar_analista()
    elif opcion == "5":
        print("Saliendo del sistema...")
        break
    else:
        print("❌ Opción inválida\n")
```
